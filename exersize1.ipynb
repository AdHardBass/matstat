```python
import numpy

from matplotlib import pyplot
import matplotlib
```

```python
def est1(arr, n): #x1 = 2 * sum(Xi) / n
    res = 0
    for i in range(n):
        res += arr[i]
    res = 2 * res / n
    return res
```

```python
def est2(arr, n): #x2 = (n + 1) * Xmin
    res = arr[0]
    for i in range(n):
        if res > arr[i]:
            res = arr[i]
    res = (n + 1) * res
    return res
```

```python
def est3(arr, n): #x3 = (n + 1) * Xmax / n
    res = arr[0]
    for i in range(n):
        if res < arr[i]:
            res = arr[i]
    res = (n + 1) * res / n
    return res
```

```python
def est4(arr, n): #x4 = Xmin + Xmax
    min_el = arr[0]
    max_el = arr[0]
    for i in range(n):
        if min_el > arr[i]:
            min_el = arr[i]
        if max_el < arr[i]:
            max_el = arr[i]
    return min_el + max_el
```

```python
def est5(arr, n): #x5 = X1 + (X2 + Xn)/(n - 1)
    if n == 1: return arr[0]
    res = 0
    for i in range(n - 1):
        res += arr[i + 1]
    res = arr[0] + res / (n - 1)
    return res
```

```python
A = 10
N = 100
arr = numpy.random.rand(N) * A
arr
```

```python
dpi = 100
fig = pyplot.figure(dpi = dpi, figsize = (1024 / dpi, 720 / dpi) )
matplotlib.rcParams.update({'font.size': 10})
pyplot.axis([0, N, 0, 10])

xs = numpy.linspace(1, N, N)
e1 = numpy.array([est1(arr, i + 1) for i in range(N)])
e2 = numpy.array([est2(arr, i + 1) for i in range(N)])
e3 = numpy.array([est3(arr, i + 1) for i in range(N)])
e4 = numpy.array([est4(arr, i + 1) for i in range(N)])
e5 = numpy.array([est5(arr, i + 1) for i in range(N)])

pyplot.plot(xs, abs(e1 - A), color = 'blue', linestyle = 'solid', label = 'e1')
pyplot.plot(xs, abs(e2 - A), color = 'green', linestyle = 'solid', label = 'e2')
pyplot.plot(xs, abs(e3 - A), color = 'red', linestyle = 'dashed', label = 'e3')
pyplot.plot(xs, abs(e4 - A), color = 'black', linestyle = 'solid', label = 'e4')
pyplot.plot(xs, abs(e5 - A), color = 'black', linestyle = 'dashed', label = 'e5')
_ = pyplot.legend(loc = 'upper right')
```
